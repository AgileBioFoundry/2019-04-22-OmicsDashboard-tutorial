# We now upload the dataset to AgileBioCyc

## Upload the  tab-delimited file

Go the `Operations menu` on the right-hand side and select [SmartTable from Uploaded file...](http://agilebiocyc:1555/groups?tab=PUBLIC&orgid=ARA#)

* Choose `all_data.tab` as the file
* Check  `Choose objects of type`
* Select `genes` from the radio buttons
* Check `Try to make objects for the first column only`
* Check `Only add rows whose first cell resolves to an object`
* Check `First line is column headers`
* Click `Upload and create SmartTable`


![upload_SmartTable.png](Images/upload_SmartTable.png)

**Disclaimer**: of the 33,341 genes in the dataset, only 3727 are represented in `AraCyc`.  AraCyc was obtained from [The Plant Metabolic Network](https://plantcyc.org) and their PGDBs only contain metabolic genes. All PGDBs available at BioCyc contain all genes in the genome.

## Rename the SmartTable you have uploaded

The title of your SmartTable has been autogenerated.  Click on the title to rename it.

![Rename smart table](Images/Rename-smart-table.png)

## Add a property column for Accession numbers

Select `Add Property->Accession-1`

![Add Accession property](Images/select-accession-property.png)

## Add a transformation column for Proteins

Select `Add Transformation->Protein Complexes`
    
![Add Protein Transformation column](Images/select-protein-transform.png)

## Move accession and protein columns to the left margin

The accession property column and the protein transformation column are all the way on the right hand side, where it is a bit inconvenient.  

### Select the `Accession-1` column 
Scroll all the way to the right and select the `Accession-1` column by clicking on the blue cell containing the `Accession-1` column header (**Be careful!**: clicking on the `Accession-1` text will make that column name editable, which is not what we want.  We want to select the column, not rename it.)
You may also have to temporarily `Hide` the `Operations menu` in order to select the `Accession-` column.

### Move the `Accession-1` column and the `Gene products` column to the left of the `locus` column

* Go to the `Operations menu` on the right hand side, and select `Column->Move->Move Column Full Left`
* Select the `Gene products` column
* Go to the `Operations menu` on the right hand side, and select `Column->Move->Move Column Full Left`
* Scroll to the left to see the result of your operations and select the `locus` column
